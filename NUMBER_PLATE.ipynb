```python
import requests
from PIL import Image
from io import BytesIO
from ultralytics import YOLO
import easyocr
import numpy as np

# Load trained YOLO model
model = YOLO("runs/detect/license_plate_detector2/weights/best.pt")

# Load image from URL
image_url = "https://img.etimg.com/thumb/width-1600,height-900,imgsize-102274,resizemode-75,msid-117001251/news/india/mumbai-two-cars-with-same-number-plate-found-outside-taj-mahal-hotel.jpg"
response = requests.get(image_url)
img = Image.open(BytesIO(response.content)).convert("RGB")

# Run YOLOv8 detection
results = model.predict(img, conf=0.5)

# Initialize OCR reader
reader = easyocr.Reader(['en'])

# Loop through all detected plates
for result in results:
    boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding boxes
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)

        # Crop the license plate region from the image
        plate_crop = img.crop((x1, y1, x2, y2))

        # Convert cropped image to array for OCR
        plate_array = np.array(plate_crop)

        # Perform OCR
        ocr_result = reader.readtext(plate_array)

        # Print OCR result
        print(f"🔍 Plate {i+1} Text:", [text[1] for text in ocr_result])
```
